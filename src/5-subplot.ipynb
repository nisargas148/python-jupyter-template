### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import numpy as np
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line plot (Top-left)
x = np.linspace(0, 10, 100)
y = x ** 2
axes[0, 0].plot(x, y, label='y = x^2', color='b')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X values')
axes[0, 0].set_ylabel('Y values')
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar plot (Top-right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['r', 'g', 'b', 'y']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter plot (Bottom-left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='b', label='Data Points')
for i in range(len(x_scatter)):
    axes[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X values')
axes[1, 0].set_ylabel('Y values')
axes[1, 0].legend()

# Pie chart (Bottom-right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'yellow'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
